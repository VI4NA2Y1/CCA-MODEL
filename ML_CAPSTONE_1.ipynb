```python
from sklearn.metrics import r2_score
number_of_observations = len(y_test)
number_of_predictors = 39  # Replace with your actual number of predictors

r_squared = r2_score(y_test, y_pred)
adjusted_r_squared = 1 - (1 - r_squared) * (number_of_observations - 1) / (number_of_observations - number_of_predictors - 1)

print("R-squared:", r_squared)
print("Adjusted R-squared:", adjusted_r_squared)
```
